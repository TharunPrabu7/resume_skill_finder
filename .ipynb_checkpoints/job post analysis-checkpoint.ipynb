```python
from collections import Counter

def find_skills(skill_set_file, skills_to_find):
    # Read the skill set from the file
    with open(skill_set_file, 'r') as file:
        skill_set = file.read()

    # Convert the skill set to lowercase
    skill_set = skill_set.lower()

    # Split the skill set by comma and strip whitespace
    skill_list = [skill.strip() for skill in skill_set.split(',')]

    # Convert skills to find to lowercase
    skills_to_find = [skill.lower() for skill in skills_to_find]

    # Count the occurrences of each skill
    skill_counts = Counter(skill_list)

    # Print the counts of each skill
    for skill in skills_to_find:
        count = skill_counts.get(skill, 0)
        if skill in skill_list:
            print(f"Skill '{skill}' found {count} times in the skill set.")

# Define the path to the skill set file
skill_set_file = 'skill_set.txt'
```

```python
# Function to extract skills from job description
def extract_skills(job_description):
    
    skill_set_file = 'skill_set.txt'
    
    # Extract skills from job description
    extracted_skills = set()
    
    # Read skills from the provided text file
    with open(skill_set_file, 'r') as file:
        skill_set = file.read()

    # Convert the skill set to lowercase
    skill_set = skill_set.lower()

    # Split the skill set by comma and strip whitespace
    skill_list = [skill.strip() for skill in skill_set.split(',')]

    # Tokenize job description using NLTK's word_tokenize
    skills_to_find = nltk.word_tokenize(job_description)
    skills_to_find = [word.lower() for word in skills_to_find]

    # Concatenate words in skills_to_find into a single string with no spaces
    paragraph = ''.join(skills_to_find)

    # Check if each skill in the skill list is present in the paragraph
    for skill in skill_list:
        # Remove spaces from the skill
        skill_no_spaces = re.sub(r'\s', '', skill)
        if skill_no_spaces in paragraph:
            extracted_skills.add(skill)

    return extracted_skills

# Define the job description
job_description = """
We are looking for a Data Scientist with expertise in Python and machine learning. 
The ideal candidate should have strong skills in data analysis, statistics, and data visualization. 
Excellent communication and teamwork abilities are also essential for this role.
"""

# Extract skills from job description
job_skills = extract_skills(job_description)

print("Skills extracted from job description:", job_skills)
```

```python
job_skills = extract_skills(example)

print("Skills extracted from job description:", job_skills)
```

```python
import PyPDF2
import docx

# Function to extract text from PDF
def extract_text_from_pdf(pdf_file):
    text = ""
    with open(pdf_file, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text()
    return text

# Function to extract text from DOC
def extract_text_from_doc(doc_file):
    text = ""
    doc = docx.Document(doc_file)
    for paragraph in doc.paragraphs:
        text += paragraph.text + "\n"
    return text

# Example usage
pdf_file = 'Tharun prabu Resume.pdf'
doc_file = 'Tharun prabu Resume.docx'

pdf_text = extract_text_from_pdf(pdf_file)
doc_text = extract_text_from_doc(doc_file)

print("Text extracted from PDF:")
print(pdf_text)

print("\nText extracted from DOC:")
print(doc_text)
```
